```python
#открываем его
with open("document.txt", "r", encoding="utf-8") as file:
    document = file.read()
```

```python
import nltk # Загрузка необходимых ресурсов NLTK
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
#Определяем класс TextProcessor
class TextProcessor:
    def __init__(self, text):
        """
        Инициализация класса с исходным текстом.
        """
        self.text = text # Сохраняем исходный текст в переменной
        self.tokens = [] # Переменная для хранения токенов после токенизации
        self.cleaned_tokens = [] #Переменная для хранения обработанных токенов (лемматизированных и очищенных)

## Метод токенизации ##
    def tokenize(self):
        """
        Метод для токенизации текста.
        """
        self.tokens = word_tokenize(self.text) # Разбиваем текст на отдельные(токены)
        return self.tokens # Возвращаем список токенов

## Метод лемматизации ##
    def lemmatize(self):
        """
        Метод для лемматизации токенов.
        """
        lemmatizer = WordNetLemmatizer() # Создаем объект лемматизатора
        self.cleaned_tokens = [lemmatizer.lemmatize(word) for word in self.tokens] # Список выполняет лемматизацию с помощью цикла, который проходится по всем токенам и возвращает его лемму, и формирует новый список лемматизированных слов
        return self.cleaned_tokens # Возвращаем список лемматизированных слов

## Метод удаления стоп-слов ##
    def remove_stopwords(self):
        """
        Метод для удаления стоп-слов.
        """
        stop_words = set(stopwords.words('english')) # Загружаем список стоп-слов и преобразуем список в множество
        self.cleaned_tokens = [word for word in self.cleaned_tokens if word not in stop_words] # С помощью цикла проходимся по словам и проверяем входят ли они в список стоп-слов, если оно есть в списке, то удаляется из него
        return self.cleaned_tokens # Возвращаем список токенов
```

```python
processor = TextProcessor(document) # Создаём объект класса TextProcessor

processor.tokenize() # С помощью этого метода разбиваем текст на токены
processor.remove_stopwords() # С помощью этого метода удаляем все стоп-слова
processor.lemmatize() # С помощью этого метода возвращаем слова к их лемме


# Вывод результата
print("Токены:", processor.tokens)
print("Лемматизированные токены:", processor.cleaned_tokens)
```
